# 레코드판 게임 (노래 맞추기 게임)

```python
import json
import random
import time
import sys
import os

try:
    print('requests 라이브러리가 있는지 확인합니다.')
    import requests
except:
    print('requests 라이브러리를 설치합니다.')
    time.sleep(1)
    !pip install requests
else:
    print('확인 되었습니다.')
    pass

class SONG_GAME:                                             
    def __init__(self):
        while 1:
            try:
                self.yourLife = int(input("당신의 주량은 몇 잔인가요? : "))
                self.p1_Life = int(input("컴퓨터1의 주량은 몇 잔인가요? : "))
                self.p2_Life = int(input("컴퓨터2의 주량은 몇 잔인가요? : "))
                self.p3_Life = int(input("컴퓨터3의 주량은 몇 잔인가요? : "))
                if self.yourLife<1 or self.p1_Life<1 or self.p2_Life<1 or self.p3_Life<1:
                    print('술을 못마시는 사람이 있습니다. 당장 귀가조치 시키세요.')
                    sys.exit()
                else:
                    break
            except ValueError:
                print('주량은 숫자로 입력해 주세요\n')            
        print()   
        self.random_guess = random.randint(2, 4)             # 컴퓨터를 틀리게 하기 위함
        self.actual_guess = 0
        self.singer_list = ['김범수','김광석','이문세','스탠딩 에그','성시경','비투비','프로미스나인','블랙핑크','방탄소년단','아이유','빅뱅','소녀시대','트와이스','레드벨벳']
        self.Lifes = [self.yourLife, self.p1_Life, self.p2_Life, self.p3_Life]   #입력된 목숨(주량)
        self.order = [0, 1, 2, 3]                            #시작 순서 주사위
        random.shuffle(self.order)
        
    def main_game(self):
        i = 0
        players = ['당신', '컴퓨터1', '컴퓨터2', '컴퓨터3']
        print("""
=================================================================================
                       레코드판 삐끼♬  레코드판 삐끼♬
                       
                  지금부터 노래 맞추기 게임을 시작하겠습니다.
                     시작 플레이어가 가수를 이야기하고,
                  나머지는 그 가수의 노래를 맞추면 됩니다.
                  게임 시작 플레이어는 무작위로 정해집니다.
        
    ★☆★☆★ 정답 기준은 '멜론'이며, 띄어쓰기에 유의하기 바랍니다  ★☆★☆★
        
=================================================================================
        
        """)
        print('\n자, 게임을 시작합니다.', end="\n\n\n")
        time.sleep(3)
        print(f'컴퓨터식 주사위!! >>>> {self.order}', end = '\n')
        time.sleep(2)
        print(f'[{players[self.order[0]]}] 이/가 게임을 시작합니다.', end="\n")
        time.sleep(3)
                            
        if self.order[0] == 0:                     # 플레이어 턴
            self.song_game_human() 
        else: 
            self.song_game_com()
        
        break_signal = 0
        while True:
            turn = self.order[i]
            if turn == 0:                          # 플레이어 턴
                answer = input("▶ 노래 제목을 말해주세요! : ")
                time.sleep(1.5)
            else:                                  # 컴퓨터 턴
                if break_signal == 1:
                    answer = "모르겠어 ㅠㅡㅠ"
                    time.sleep(1.5)
                else: 
                    answer = random.choice(self.song_list)
                print("▶ {}의 답 : {}".format(players[turn], answer))
                time.sleep(1.5)
            
            if answer in self.song_list:
                print("오케이 정답!", end = "\n\n")
                time.sleep(1)
                del self.song_list[self.song_list.index(answer)]
            else:
                print("어어라?? 마시면서 배우는 술게임~", end = "\n\n")
                break

            i += 1
            i = i % 4
            
            self.actual_guess += 1
            if self.actual_guess >= self.random_guess:
                break_signal = 1
                

        print(f'{players[turn]} 벌주 드가자~~~~~~')
        time.sleep(1)
        print('술이 들어간다!')
        time.sleep(1)
        print('쭉 쭉쭉 쭉쭉~♪')
        time.sleep(1)
        print('쭉 쭉쭉 쭉쭉~♪')
        time.sleep(1)
        print('언제까지 어깨춤을 추게할거야!')
        time.sleep(1)
        print('내어깨를 봐!! 탈골됐잖아!!! 탈골!  탈골!  탈골!탈골!탈골!탈골!\n')
        time.sleep(2)
        self.Lifes[turn] -= 1
        return self.Lifes


    def load_song_list(self, singer):                    # 멜론에서 곡 정보들 받아오기
        url = "https://www.melon.com/search/keyword/index.json" # 멜론에서 검색할 떄 생성되는 Json 파일의 url 따옴
        params = {
            'jscallback' : "jQuery191034735473490667346_1644218622270", # 쿼리문을 나눠서 기입, 쿼리(검색어)는 들어오는 인수로 처리
            'query' : singer}
        headers = {"User-Agent" : "Chrome/75.0.3770.80" } # 파이썬 스크립트를 통해 접속하는 것을 크롬을 통해 한 것으로 속임
                                                          # 파이썬 스크립트로 하면 접속을 안해줌..
        response = requests.get(url, headers = headers, params = params)
        response = response.text

        json_string = response.replace(params['jscallback'] + '(', '').replace(');', '') # 
        result_dict = json.loads(json_string)
        return result_dict


    def song_game_human(self):
        while True:                                    # 가수 이름을 잘못 입력할 것을 대비
            singer = input('\n가수를 입력하세요 : ')
            time.sleep(1)
            song_dict = self.load_song_list(singer)    # 크롤링으로 얻은 dictionary
            if 'SONGCONTENTS' not in song_dict.keys():
                print("그런 가수가 어딨어 ㅋㅋㅋ")
                print('가수 이름 다시 입력하세요')
            elif len(song_dict['SONGCONTENTS']) == 0:
                print("마 취했나; 똑디 말해라")
                print('가수 이름 다시 입력하세요')
            else:
                break 
        self.song_list = self.get_song_list(singer, song_dict)   #가수 검색 결과


    def get_song_list(self, singer, song_dict):
        song_list = []
        for song in song_dict["SONGCONTENTS"]:
            tmp = song['SONGNAME'].split("(")     # 노래 제목에 괄호 대비
            song_list.append(tmp[0].strip())      # 노래 제목 뒤 띄어쓰기 제거
        return song_list      

    
    def song_game_com(self):
        singer = random.choice(self.singer_list)
        print(f"\n플레이어가 선택한 가수는 [ {singer} ]입니다.", end = '\n\n')
        print(f'▶ [ {singer} ]의 노래를 말해주세요.', end = '\n\n\n')
        time.sleep(2)
        song_dict = self.load_song_list(singer)
        self.song_list = self.get_song_list(singer, song_dict) #컴퓨터가 랜덤하게 뽑은 가수를 검색한 결과 리스트

        
    def EndOrRe(self,Lifes):     # 반복시킬 떄 게임을 끝낼지 말지와 메세지를 위함
        try:
            for i in Lifes:
                k = int(i)
        except:
            print('●주량이 잘못 입력되었습니다. 프로그램을 종료합니다.')
            sys.exit()
        while 1:
            if Lifes[0] == 0:
                print('\n당신이 뻗었습니다! 게임을 종료합니다.')
                break
            elif Lifes[1] == 0:
                print('\n[컴퓨터1]이 뻗었습니다! 게임을 종료합니다.')
                break
            elif Lifes[2] == 0:
                print('\n[컴퓨터2]이 뻗었습니다! 게임을 종료합니다.')
                break
            elif Lifes[3] == 0:
                print('\n[컴퓨터3]이 뻗었습니다! 게임을 종료합니다.')
                break
            else:
                print(f"""
            ●남은 주량 체크! 과음은.. 아주 좋아요~
    당신: {Lifes[0]}잔, [컴퓨터1]: {Lifes[1]}잔, [컴퓨터2]: {Lifes[2]}잔, [컴퓨터3]: {Lifes[3]}잔,          
                """)
                print('★다들 아직 쌩쌩합니다. 다음 게임을 진행하겠습니다★')
                self.main_game()
```

```python
# 게임 시작 및 진행
game = SONG_GAME()              # 클래스 호출, 객체 생성
new_life = game.main_game()     # 한 게임이 돌고 난 뒤에 남은 목숨들을 반환해줌
game.EndOrRe(new_life)          # 남은 목숨을 이용해 게임을 끝낼지 더 진행할지 판단 후 추가 진행
```
